```python
import requests as req
from bs4 import BeautifulSoup

page_count = 1

while True:
    URL = f'https://quotes.toscrape.com/page/{page_count}/'
    res = req.get(URL)

    soup = BeautifulSoup(res.text, 'lxml')
    quotes = soup.select('div.quote')

    if not quotes:
        print('no valid pages anymore...')
        break

    with open(f'scraped_data2/quotes{page_count}.html', 'w', encoding='utf-8') as f:
        f.write(res.text)
        print(f'downloaded data from page{page_count}')
    page_count += 1 
```

```python
all_data = [] 
for page_no in range(1, 11):
    with open(f'scraped_data2/quotes{page_no}.html', 'r', encoding='utf-8') as f:
        html_content = f.read()

    soup = BeautifulSoup(html_content, 'lxml')
    quotes = soup.select('div.quote')
    for q in quotes:
        all_tags = q.select('.tags .tag')
        for tag in all_tags:
            if tag.get_text() == 'life':
                quote  = q.select_one('.text').get_text()   
                author = q.select_one('.author').get_text() 
                all_data.append([quote, author])

all_data
```

```python
import pandas as pd

df = pd.DataFrame(all_data, columns = ['Quote', 'Author'])
df.to_csv('author_data.csv')
```
