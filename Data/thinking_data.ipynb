```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
    return data
```

```python
data = load_data('store_data.json')
print(data)
print(type(data))
```

```python
#clean and structure the data
def clean_data(data):
    text_to_num = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
    cleaned_data = []
    unique_user = set()
    for user in data:
        #clean rating
        raw_rating = str(user['rating']).strip().lower()
        if raw_rating in text_to_num:
            user['rating'] = text_to_num[raw_rating]

        #handle missing values
        raw_age = user.get('age')
        if raw_age == None:
            user['age'] = None

        #deduplication
        if user['name'] in unique_user:
            continue
        cleaned_data.append(user)
        unique_user.add(user['name'])

    print(cleaned_data)
    return cleaned_data
        
```

```python
data = clean_data(data)
```

```python
#get meaningful insights from data
def get_insights(data):
    #avg rating
    tot_rating = 0
    for user in data:
        tot_rating += float(user['rating'])
    print('avg rating is: ', tot_rating / len(data))

    #percentage of users with poor rating
    poor_ratings = 0
    for user in data:
        if float(user['rating']) < 3:
            poor_ratings += 1
    print(f'percentage of poor rating is: {(poor_ratings/len(data)) * 100}%')
```

```python
get_insights(data)
```

```python
#recommendation system
def get_recommendations(data):
    recommendations = []
    for user in data:
        curr_rec = {}
        curr_rec['name'] = user['name']
        if float(user['rating']) >= 4:
            curr_rec['brand'] = 'Apple'
        else:
            curr_rec['brand'] = 'Samsung'
        recommendations.append(curr_rec)
    print(recommendations)
```

```python
get_recommendations(data)
```
